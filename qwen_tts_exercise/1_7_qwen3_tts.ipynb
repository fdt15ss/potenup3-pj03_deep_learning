# 1. 커스텀 보이스 생성

```python
import torch 
from qwen_tts import Qwen3TTSModel 

device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(f"사용 디바이스: {device}")
```

## 1) 모델 불러오기

```python
model = Qwen3TTSModel.from_pretrained(
    "Qwen/Qwen3-TTS-12Hz-0.6B-CustomVoice",  # 가장 가벼운 모델부터
    device_map=device,
    dtype=torch.bfloat16,
    attn_implementation="sdpa",  # Windows는 sdpa 고정
)
```

## 2) 가능 목록 출력하기

```python
# 지원 화자/언어 확인
print("화자 목록:", model.get_supported_speakers())
print("언어 목록:", model.get_supported_languages())
```

## 3) TTS

```python
import soundfile as sf
# 한국어 테스트
wavs, sr = model.generate_custom_voice(
    text="안녕하세요, 저는 인공지능 스피커입니다.",
    language="Korean",
    speaker="sohee",  # 목록 확인 후 한국어 지원 화자로 변경
)

sf.write("./audio/qwen3_test.wav", wavs[0], sr)
print("✅ test.wav 생성 완료!")
```

```python
from IPython.display import Audio, display 

display(Audio("./audio/qwen3_test.wav", autoplay=True))
```

# 2. 보이스 디자인

```python
import torch
import soundfile as sf
from qwen_tts import Qwen3TTSModel
from IPython.display import Audio, display

device = "cuda:0" if torch.cuda.is_available() else "cpu"

model = Qwen3TTSModel.from_pretrained(
    "Qwen/Qwen3-TTS-12Hz-1.7B-VoiceDesign",  # VoiceDesign 모델
    device_map=device,
    dtype=torch.bfloat16,
    attn_implementation="sdpa",
)

wavs, sr = model.generate_voice_design(
    text="안녕하세요, 오늘 날씨가 정말 좋네요!",
    instruct="따뜻하고 친근한 70대 여성 목소리, 밝은 톤으로",  # 목소리 묘사
)

sf.write("./audio/voice_design.wav", wavs[0], sr)
display(Audio("./audio/voice_design.wav", autoplay=True))
```

# 3. 보이스 클론

```python
import torch
import soundfile as sf
from qwen_tts import Qwen3TTSModel
from IPython.display import Audio, display

device = "cuda:0" if torch.cuda.is_available() else "cpu"

model = Qwen3TTSModel.from_pretrained(
    "Qwen/Qwen3-TTS-12Hz-0.6B-Base",  # Base 모델
    device_map=device,
    dtype=torch.bfloat16,
    attn_implementation="sdpa",
)
```

```python

wavs, sr = model.generate_voice_clone(
    text="안녕하세요, 이 목소리는 복제된 목소리입니다.",
    language="Korean",
    ref_audio="./audio/my_voice.mp3",        # 복제할 참조 음성 파일
    ref_text="안녕하세요. 개그우먼 이수지입니다. 반갑습니다.",  # 참조 음성의 스크립트
)

sf.write("./audio/voice_clone.wav", wavs[0], sr)
display(Audio("./audio/voice_clone.wav", autoplay=True))
```
