```python
import requests
from pprint import pprint

url = "http://localhost:11434/api/generate"
data = {"model": "deepseek-r1:8b", "prompt": "네 소개를 200자 이내로 해줘."}
```

```python
import json

thinking = ""
response = ""

with requests.post(url=url, json=data) as r:
    for line in r.iter_lines():
        if line:
            data = json.loads(line.decode("utf-8"))

            thinking += data.get("thinking", "")
            response += data.get("response", "")

print("[Thinking]")
print(thinking)
print("[Response]")
print(response)
```
